```python
hotels_cont_1 = [] 
cont_index = 8
num_countries = dat['geo'][cont_index]['countries']['pagination']['total']

for i in range(num_countries):
    num_destinations = dat['geo'][cont_index]['countries']['countriesPage'][i]['destinations']['pagination']['total']
    for ii in range(num_destinations):
        num_resorts = dat['geo'][cont_index]['countries']['countriesPage'][i]['destinations']['destinationsPage'][ii]['resorts']['pagination']['total']
        for iii in range(num_resorts):
            num_products = dat['geo'][cont_index]['countries']['countriesPage'][i]['destinations']['destinationsPage'][ii]['resorts']['resortsPage'][iii]['products']['pagination']['total']
            for iiii in range(num_products):
                hotel = dat['geo'][cont_index]['countries']['countriesPage'][i]['destinations']['destinationsPage'][ii]['resorts']['resortsPage'][iii]['products']['productsPage'][iiii]
                hotels_cont_1.append(hotel)
```
